### Задание 1

```python
"""
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. 
Определите какому фильму было выставлено больше всего оценок 5.0.
"""
```

```python
import pandas as pd
```

```python
# Загружаем файл с фильмами
mov_df = pd.read_csv('ml-latest-small/movies.csv')
mov_df.head(10)
```

```python
# Загружаем файл с рейтингом
rat_df = pd.read_csv('ml-latest-small/ratings.csv')
rat_df.head(10)
```

```python
# Объединяем таблицы 
current_df = mov_df.merge(rat_df, left_on = 'movieId', right_on = 'movieId', how = 'inner')
current_df.head()
```

```python
# Выводим фильмы с максимальным количеством оценок "5"
film_max_rat = (current_df
 .query('rating == 5.0')
 .groupby('title')
 .agg(count_rat = ('rating', 'sum'))
 .sort_values('count_rat', ascending=False)
)
film_max_rat.head()
```

```python
# Получаем результат
best_movie = film_max_rat.loc[film_max_rat['count_rat'] == 765.0]
print(best_movie)
```

### Задание 2

```python
"""
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики 
(Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. 
Не учитывайте в расчетах отрицательные значения quantity.
"""
```

```python
# Загружаем датафрейм с потреблением по странам
country_df = pd.read_csv('power.csv')
country_df.head(50)
```

```python
# Получаем потребление по странам Прибалтики
balt_df = (country_df
 .query('category in (4, 12, 21)')
 .query('country in ("Estonia", "Latvia", "Lithuania")')
 .query('quantity >= 0')
 .query('year in (2005, 2006, 2007, 2008, 2009, 2010)')
 .groupby('country')
 .agg(sum_quantity = ('quantity','sum'))          
)
balt_df.head(50)
```

```python
print('Cуммарное потребление стран Прибалтики: {}'.format(sum(balt_df['sum_quantity'])))
```
